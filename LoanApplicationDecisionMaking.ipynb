## Cleaning

```python
df2 = df.copy()
```

```python
# Drop irrelevant columns
df2.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis = 1, inplace=True)
```

```python
# Fill all the missing categorical variable with "Unknown"
df2["Credit_Card_types"].fillna("Unknown", inplace=True)
df2["Property_Type"].fillna("Unknown", inplace=True)
```

```python
# Fill all the remaining missing data with the median of its respective column
df2 = df2.fillna(df2.median())
```

```python
# Check the unique values for each categorical columns
print(df2.Employment_Type.unique())
print(df2.Property_Type.unique())
print(df2.State.unique())
```

```python
# A few states are same but written differently
# Change all the duplicate states to be the same
df2['State'] = df2['State'].replace(['P.Pinang','Pulau Penang'],'Penang')
df2['State'] = df2['State'].replace(['Johor B'],'Johor')
df2['State'] = df2['State'].replace(['K.L'],'Kuala Lumpur')
df2['State'] = df2['State'].replace(['N.S'],'N.Sembilan')
df2['State'] = df2['State'].replace(['SWK'],'Sarawak')
```

## Label Encoding

```python
# Label Encode all the categorical variables
from sklearn.preprocessing import LabelEncoder
df3 = df2.copy()
le = LabelEncoder()
categorical_feature_mask = df3.dtypes==object
categorical_cols = df3.columns[categorical_feature_mask]
df3[categorical_cols] = df3[categorical_cols].apply(lambda col: le.fit_transform(col))
```

## Feature Engineering

```python
X_resampled["Loan_Amount_Annually"] = X_resampled["Loan_Amount"] / X_resampled["Loan_Tenure_Year"]
```

```python
X_resampled.head()
```

```python
# Normalize the variables with big values for faster processing
X_scaled = X_resampled.copy()
X_scaled["Loan_Amount"] = (X_resampled["Loan_Amount"]-X_resampled["Loan_Amount"].min(axis=0)) / ((X_resampled["Loan_Amount"].max(axis=0)) - X_resampled["Loan_Amount"].min(axis=0))
X_scaled["Monthly_Salary"] = (X_resampled["Monthly_Salary"]-X_resampled["Monthly_Salary"].min(axis=0)) / ((X_resampled["Monthly_Salary"].max(axis=0)) - X_resampled["Monthly_Salary"].min(axis=0))
X_scaled["Total_Sum_of_Loan"] = (X_resampled["Total_Sum_of_Loan"]-X_resampled["Total_Sum_of_Loan"].min(axis=0)) / ((X_resampled["Total_Sum_of_Loan"].max(axis=0)) - X_resampled["Total_Sum_of_Loan"].min(axis=0))
X_scaled["Loan_Amount_Annually"] = (X_resampled["Loan_Amount_Annually"]-X_resampled["Loan_Amount_Annually"].min(axis=0)) / ((X_resampled["Loan_Amount_Annually"].max(axis=0)) - X_resampled["Loan_Amount_Annually"].min(axis=0))
X_scaled["Total_Income_for_Join_Application"] = (X_resampled["Total_Income_for_Join_Application"]-X_resampled["Total_Income_for_Join_Application"].min(axis=0)) / ((X_resampled["Total_Income_for_Join_Application"].max(axis=0)) - X_resampled["Total_Income_for_Join_Application"].min(axis=0))
```

```python
#!pip install boruta
from sklearn.ensemble import RandomForestClassifier
from boruta import BorutaPy
from sklearn.preprocessing import MinMaxScaler
```

```python
def ranking(ranks, names, order=1):
    minmax = MinMaxScaler()
    ranks = minmax.fit_transform(order*np.array([ranks]).T).T[0]
    ranks = map(lambda x: round(x,2), ranks)
    return dict(zip(names, ranks))
```

```python
# newX = X_scaled.loc[:, X_scaled.columns != 'Decision']
# newy = X_scaled.loc[:, X_scaled.columns == 'Decision']

# colnames = newX.columns
# print(colnames)

newX = X_scaled
newy = y_resampled

colnames = newX.columns
#print(colnames)
```

```python
rf = RandomForestClassifier(n_jobs=-1, class_weight='balanced', max_depth=5)
feat_selector = BorutaPy(rf, n_estimators="auto", random_state = 1)
feat_selector.fit(newX.values, newy.values.ravel())

boruta_score = ranking(list(map(float, feat_selector.ranking_)), colnames, order=-1)
boruta_score = pd.DataFrame(list(boruta_score.items()), columns=['Features', 'Score'])
```

# Machine Learning

## Association Rule Mining Using Apriori Algorithm

```python
# Get all the object variables
categorical_feature_mask = df2.dtypes==object
categorical_cols = df2.columns[categorical_feature_mask]
df_objects = df2[categorical_cols] 
df_objects = df_objects.fillna('Unknown')
```

```python
df_oneHot = pd.get_dummies(df_objects, columns=['Employment_Type'])
df_oneHot = pd.get_dummies(df_oneHot, columns=['More_Than_One_Products'])
df_oneHot = pd.get_dummies(df_oneHot, columns=['Credit_Card_types'])
df_oneHot = pd.get_dummies(df_oneHot, columns=['Property_Type'])
#df_oneHot = pd.get_dummies(df_oneHot, columns=['State'])
df_oneHot = pd.get_dummies(df_oneHot, columns=['Decision'])
```

```python
df_oneHot
df_oneHot.drop("State", axis = 1, inplace=True)
#dropping state because State as a consequent/antecedent possibly does not make sense for assoc rules
```

```python
# Find frequently occuring itemsets
from mlxtend.frequent_patterns import apriori
frequent_itemsets_ap = apriori(df_oneHot, min_support=0.01, use_colnames=True)
```

```python
print(frequent_itemsets_ap)
```

```python
# Find frequently occuring objects using F-P growth
from mlxtend.frequent_patterns import fpgrowth

frequent_itemsets_fp=fpgrowth(df_oneHot, min_support=0.01, use_colnames=True)
```

```python
print(frequent_itemsets_fp)
```

```python
# Association Rule Mining
from mlxtend.frequent_patterns import association_rules

rules_ap = association_rules(frequent_itemsets_ap, metric="confidence", min_threshold=0.8)
```

```python
rules_ap.consequents.unique()
```

```python
rules_ap
```

```python
rules_ap2 = rules_ap.drop(rules_ap.loc[rules_ap['consequents']!=frozenset({'Decision_Accept'})].index, inplace=False)
#removed all rows which did not result in Decision_Accept as a consequent
```

```python
pd.options.display.width = 100
pd.options.display.max_colwidth = 150
#increased width for easier reading
```

```python
#displayed all antecedents which resulted in Decision_Accept as a consequent, in descending order of confidence
rules_ap2[['antecedents', 'confidence']].sort_values(by=['confidence'], ascending=False)
```

## K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
df_kmeans = X_features.copy()
```

```python
plt.scatter(df_kmeans.Monthly_Salary, df_kmeans.Loan_Amount)
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df_kmeans[["Monthly_Salary", "Loan_Amount"]])
    sse.append(km.inertia_)
```

```python
plt.xlabel("K")
plt.ylabel("Sum of squared error")
plt.plot(k_rng,sse)
```

```python
km = KMeans(n_clusters=4)
```

```python
y_predicted = km.fit_predict(df_kmeans[["Monthly_Salary", "Loan_Amount"]])
```

```python
df_kmeans["cluster"] = y_predicted
c1 = df_kmeans[df_kmeans.cluster==0]
c2 = df_kmeans[df_kmeans.cluster==1]
c3 = df_kmeans[df_kmeans.cluster==2]
c4 = df_kmeans[df_kmeans.cluster==3]

plt.scatter(c1.Monthly_Salary, c1.Loan_Amount, color="green")
plt.scatter(c2.Monthly_Salary, c2.Loan_Amount, color="red")
plt.scatter(c3.Monthly_Salary, c3.Loan_Amount, color="blue")
plt.scatter(c4.Monthly_Salary, c4.Loan_Amount, color="yellow")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1], color="purple", marker="+", label="centroid")
plt.xlabel("Monthly Salary")
plt.ylabel("Loan Amount")
plt.legend()
# plt.savefig("kmeans.jpg", bbox_inches = 'tight')
```

## Classification Algorithms

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_features, y_resampled, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
```

```python
logreg = LogisticRegression(solver='lbfgs', max_iter=500)
logreg.fit(X_train,y_train.values.ravel())
y_predLOGREG=logreg.predict(X_test)
y_predLOGREG_proba = logreg.predict_proba(X_test)[::, 1]
```

```python
svmclf = svm.SVC(kernel='linear', probability=True) 
svmclf.fit(X_train, y_train.values.ravel())
y_predSVM = svmclf.predict(X_test)
y_predSVM_proba =svmclf.predict_proba(X_test)[::, 1]
```

```python
clfRFC=RandomForestClassifier(n_estimators=100)
clfRFC.fit(X_train,y_train.values.ravel())
y_predRFC=clfRFC.predict(X_test)
y_predRFC_proba =clfRFC.predict_proba(X_test)[::, 1]
```

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train.values.ravel())
y_predGNB = gnb.predict(X_test)
y_predGNB_proba =gnb.predict_proba(X_test)[::, 1]
```

```python
# #!pip install tensorflow
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(20,)),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(256, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(2)
])
```

```python
model.compile(optimizer="adamax",
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=150, batch_size=8)
```

```python
# y_predann = model.predict(X_test)
results = model.evaluate(X_test, y_test, batch_size=8)
print("test loss, test acc:", results)
```
